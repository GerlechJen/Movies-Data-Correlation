```python
# import libraries
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plots we will create
```

```python
# Read in the data 
df = pd.read_csv('./movies.csv')
df
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
df.describe()
```

```python
# Getting summary data of our columns 
df.info()
```

```python
#  Checking for missing data per column 
df.isnull().sum()
```

```python
for column in df.columns:
    nulls = df[column].isnull().sum()
    print(f'{column} - {nulls} null values')
```

```python
# df = df.dropna
```

```python
# Checking data types of our columns 
df.dtypes
```

```python
# Changing data type of budget and gross columns from float to integer 
df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')
```

```python
df
```

```python
#  Creating correct year column 
df['CorrectYear'] = df['released'].astype(str).str[:4]
```

```python
df
```

```python
df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
df
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop duplicates
df = df.drop_duplicates()
df.head()
```

```python
df.count()
```

```python
# Scatter plot with budget vrs gross
plt.scatter(x=df['budget'], y = df['gross'])
plt.title('Budget vrs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show
```

```python
# Plot the budget vrs gross using seaborn 
sns.regplot(x ='budget', y='gross', data=df, scatter = True, scatter_kws = {'color': 'red'}, line_kws = {'color': 'blue'})
```

```python
#  Looking at correlation using pearson
df.corr()
```

```python
#  Looking at correlation using kendall
df.corr(method = 'kendall')
```

```python
#  Looking at correlation using spearman
df.corr(method = 'spearman')
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#  Making the categorical columns numerical
df_numerized = df 
for column in df_numerized.columns:
    if df_numerized[column].dtype == 'object':
        df_numerized[column] = df_numerized[column].astype('category')
        df_numerized[column] = df_numerized[column].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
